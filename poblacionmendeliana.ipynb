```python
import scipy

def generacion_de_la_poblacion(N,p):
    """Esta función indica N individuos;cada individuo tiene dos 
    cromosomas que contienen el cromosoma A o a ,con probabilidad p o 1-p,La población es una lista de tuplas 
    """
    population = []
    for i in range(N):
        allele1 = "A"
        if scipy.random.rand() > p:
            alelle1 = "a"
        allele2 = "A"  
        if scipy.random.rand() > p:
            allele2 = "a"
        population.append((allele1, allele2))  
    return population      
```

```python
generacion_de_la_poblacion(N = 500,p = 0.5)
```

```python
def conteo_de_frecuencia_de_alelos(population):
    """Esta función toma una población como entrada y genera el recuento de cada alelo
    """
    AA = population.count(("A", "A"))
    Aa = population.count(("A", "a"))
    aA = population.count(("a", "A"))
    aa = population.count(("a", "a"))
    return({"AA": AA, "aa": aa, "Aa": Aa, "aA":aA})
```

```python
my_pop = generacion_de_la_poblacion(500, 0.5)
```

```python
fmy_pop = conteo_de_frecuencia_de_alelos(my_pop)
print(fmy_pop)
print((500-fmy_pop["aa"])/500)
```

```python
def reproduccion_de_la_poblacion(population):
    """Esta funcion que toma la poblacion actual y produce la proxima generacion
    """
    new_generation = []
    N = len(population)
    for i in range(N):
        #
        dad = scipy.random.randint(N)
        mom = scipy.random.randint(N)
        #
        chr_mom = scipy.random.randint(2)
        offspring = (population[mom][chr_mom], population[dad][1 - chr_mom])
        #if offspring == ("a", "a"):
          #next() 
        new_generation.append(offspring)
    return new_generation


    
```

```python
fmy_pop = conteo_de_frecuencia_de_alelos(my_pop)
print(fmy_pop)
(500 - fmy_pop["aa"])/500

```
